```r
next_state <- function(state){
    u <- runif(1)
    if (state == 0){
        if (u < .3){
            return(0)
        }
        else{
            return(1)
        }
    }
    else{
        if (u < .6){
            return(0)
        }
        else{
            return (1)
        }
    }
}
```

```r
vector_N <- numeric()


#print(state)

for (j in 1:1000){
    state <- 0
    N <- 1
    for (i in 1:10000){
        state <- next_state(state)
        #print(state)
        if(state == 0){
            N <- N + 1
        }
    }
    vector_N <- c(vector_N, N)
}
```

```r
hist(vector_N)
```

```r
png(filename = "hist.png", height = 2000, width = 2000, res = 400)
hist(vector_N, col="darkcyan", main="", xlab="Passages through state 0 up to time 10,000")
dev.off()
```
