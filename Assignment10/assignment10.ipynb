### Ex. 12 p. 264

```r
standardize <- function(vector){
    return ((vector - mean(vector))/sqrt(var(vector)))
}

means_unif <- numeric()
means_unif_std <- numeric()

var_unif <- numeric()
var_unif_std <- numeric()

for(i in 1:1000){
    x <- runif(1000)
    
    means_unif <- c(means_unif, mean(x))
    means_unif_std <- c(means_unif_std, mean(standardize(x)))
    
    var_unif <- c(var_unif, var(x))
    var_unif_std <- c(var_unif_std, var(standardize(x))) 
    
}
```

```r
plot(means_unif, col="red", ylim=c(-0.0001, .6), xlab="repetition", ylab="mean")
points(means_unif_std, col="blue")
legend(x = "right", legend = c("Original", "Standardized"), fill = c("red", "blue"), 
       title = "")
```

```r
png(filename="means_unif.png", height=2000, width=2000, res=400)
plot(means_unif, col="red", ylim=c(-0.0001, .6), xlab="repetition", ylab="mean")
points(means_unif_std, col="blue")
legend(x = "right", legend = c("Original", "Standardized"), fill = c("red", "blue"), 
       title = "")
dev.off()
```

```r
plot(var_unif, col="red", ylim=c(0, 1.1), xlab="repetition", ylab="variance")
points(var_unif_std, col="blue")
legend(x = "right", legend = c("Original", "Standardized"), fill = c("red", "blue"), 
       title = "")
```

```r
png(filename="vars_unif.png", height=2000, width=2000, res=400)
plot(var_unif, col="red", ylim=c(0, 1.1), xlab="repetition", ylab="variance")
points(var_unif_std, col="blue")
legend(x = "right", legend = c("Original", "Standardized"), fill = c("red", "blue"), 
       title = "")
dev.off()
```

### Ex. 3 p. 278

```r
library(distr)
p    <- function(x){ (0.05**2) * (x*exp(-0.05*x)) } # probability density function
dist <-AbscontDistribution(d=p)  # signature for a dist with pdf ~ p
rdist <- r(dist)                 # function to create random variates from p

mean_x <- numeric()
var_x <- numeric()

for (i in 1:1000){
    X <- rdist(1000)
    mean_x <- c(mean_x, mean(X))
    var_x <- c(var_x, var(X))
}
```

```r
mean(mean_x)
```

```r
boxplot(mean_x, col="cyan3")
```

```r
png(filename="lifetime_mean.png", height=2000, width=2000, res=400)
boxplot(mean_x, col="cyan3")
dev.off()
```

```r
boxplot(var_x, col="bisque")
```

```r
png(filename="lifetime_variance.png", height=2000, width=2000, res=400)
boxplot(var_x, col="bisque")
dev.off()
```

```r
mean(var_x)
```

### Ex. 19 p. 249

```r
answers = c('a', 'b', 'c', 'd')

scores <- numeric()

for(i in 1:10000){

    correct_answer <- sample(answers, 1)

    answerset <- sample(answers, sample(c(1,2,3,4), 1))

    if (correct_answer %in% answerset){
        score <- 3 - (length(answerset)-1)
    }else{
        score <- -length(answerset)
    }
    
    scores <- c(scores, score)
}
```

```r
boxplot(scores, col="cornflowerblue")
```

```r
png(filename="scores.png", height=2000, width=2000, res=400)
boxplot(scores, col="cornflowerblue")
dev.off()
```

```r
mean(scores)
```
