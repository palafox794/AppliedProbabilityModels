Aditionally, for both vectors, compute their minimum and maximum values, as well as the .25 and .75 quantiles. (a) For the first case, we give a vector where 52 of its entries are equal to 200 (the smallest value permitted) and 48 entries are 800 (the largest value permitted). This will give us a 200 median, and skew the average way above that.

````r
x <- seq(200, 200, length.out=52)
````

````r
y <- seq(800, 800, length.out=48)
````

````r
data1 <- c(x, y)
````

````r
mean(data1)
````

````r
median(data1)
````

````r
summary(data1)
````

(b) For the second case, we give a hundred-element vector where all its entries are equal to 201, except the first one, which equals 200. Median will obviously be 201, but since one element is less than 201 (the first one, equal to 200) then the mean must be less than 201.

````r
x <- seq(201, 201, length.out=99)
````

````r
data2 <- c(200, x)
````

````r
summary(data2)
````

````r
mean(data2)
````

````r
median(data2)
````

The boxplots:

````r
boxplot(data1, data2)
````

--------------------------
To plot the violin chart, we need the ```ggplot2``` library. Intuitively (at least from my understanding) the violin plot displays the same information as a boxplot, but instead of having a rectangle-shaped figure, you get an approximation of the density of the data.

````r
library(ggplot2)
````

We create a dataframe with the values of our first vector, and a label column. All elements get label 'A'. We do the same for the second vector, labelling its elements as B.

````r
df1 <- data.frame(name = strrep(c("A"), seq(1,1, length.out=100)), value = data1)
````

````r
df1
````

````r
df2 <- data.frame(name = strrep(c("B"), seq(1,1, length.out=100)), value = data2)
````

````r
df2
````

The following gives us the violin plot for the data in the first vector 

````r
p <- ggplot(df1, aes(x=name, y=value, fill=name)) + # fill=name allow to automatically dedicate a color for each group
  geom_violin()
````

````r
p
````

The following gives us a violin plot for the data in the second vector

````r
q <- ggplot(df2, aes(x=name, y=value, fill=name)) + # fill=name allow to automatically dedicate a color for each group
  geom_violin()
````

````r
q
````

Now we concatenate our dataframes in a single one, and create a violin plot with the same procedure. Since elements from the first and second vector can be distinguished by their label ('A' or 'B'), we can use the same procedure and ggplot will create two different violin plots in the same box.

````r
df <- rbind(df1, df2)
````

````r
p <- ggplot(df, aes(x=name, y=value, fill=name)) + # fill=name allow to automatically dedicate a color for each group
  geom_violin()
````

````r
p
````
