```r
cond_prob <- function(p, sens, sp){
    #Calculates conditional probability
    #of having covid given that you tested positive.
    #Depends on parameter p := "probability of having covid", 
    #sens := "test sensitivity" and sp := "test specificity".
    
    P <- (sens*p) / (sens*p + (1-sp)*(1-p))

    return(P)
}
```

```r
sens <- .71
sp <- .95
p_cov <- 77807/(5.4e6)
```

```r
#Covid with marginal probability = cases / population
cond_prob(p_cov, sens, sp)
```

```r
#Covid with high marginal prob
cond_prob(p=.2, sens, sp)
```

```r
x <- seq(0.015, 0.5, by = 0.0005)

y <- numeric()
for(p in x){
    y <- c(y, cond_prob(p, sens = .71, sp = .95))
}

#Sensitivity and Specificity are fixed here at 71% and 95% resp.
plot(x,y, xlab="Marginal probability of having COVID",
    ylab="Conditional probability of having COVID given a positive test")
```

```r
png(filename="covid_sens71_spec95.png", height=2000, width=2000, res=400)
plot(x,y, xlab="Marginal probability of having COVID",
    ylab="Conditional probability of having COVID given a positive test")
dev.off()
```
